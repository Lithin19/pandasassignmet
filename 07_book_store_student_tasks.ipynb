## Task 44

Use apply() to categorize Unit_Price into Low, Medium, High.

```python
def price_cat(p):
    if p < 500: return 'Low'
    if p < 1000: return 'Medium'
    return 'High'
df['Price_Category'] = df['Unit_Price'].apply(price_cat)
```
